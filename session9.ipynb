```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_nums = np.random.uniform(0, 1, 1000)
```

```python
plt.hist(random_nums, bins=100, edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Uniformly Distributed Random Numbers")
plt.grid(True)
plt.tight_layout()
plt.savefig("random_histogram.pdf")  # Saves the plot as a PDF
plt.show()
```
