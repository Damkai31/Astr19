```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()
random_expo = rng.exponential(scale=1.0, size=1000)
```

```python
plt.hist(random_expo, bins=100, edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Exponentially Distributed Random Numbers")
plt.grid(True)
plt.tight_layout()
plt.savefig("exponential_histogram.pdf")
plt.show()
```
